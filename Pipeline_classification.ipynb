```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score
from sklearn.decomposition import TruncatedSVD

# Splitting the data into train and test sets
IV = ["MSA", "MIP", "OCLTV", "DTI", "OrigUPB", "PropertyState", "EverDelinquent", "MonthsDelinquent", "LTV_Range", "CreditRange", "RepPayRange"]

X = df[IV]
Y = df[["EverDelinquent"]]  # Convert the target variable to a DataFrame

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=2)

# Reshape y_train to have two dimensions
y_train_reshaped = y_train.values.reshape(-1, len(Y.columns))

# Preprocessing transformers
numeric_features = ['MSA', 'MIP', 'OCLTV', 'DTI', 'OrigUPB', 'MonthsDelinquent']
categorical_features = ['PropertyState', 'LTV_Range', 'CreditRange', 'RepPayRange']

numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('numeric', numeric_transformer, numeric_features),
    ('categorical', categorical_transformer, categorical_features)
])

# Create separate pipelines for each target variable
pipelines = [
    Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression())
    ])
    for column in y_train_reshaped.T  # Iterate over columns of y_train_reshaped
]

# Fit the pipelines to the training data
for pipeline, target in zip(pipelines, Y.columns):
    pipeline.fit(x_train, y_train_reshaped)
    
# Fit the pipeline to the training data
pipeline.fit(x_train, y_train)

```
